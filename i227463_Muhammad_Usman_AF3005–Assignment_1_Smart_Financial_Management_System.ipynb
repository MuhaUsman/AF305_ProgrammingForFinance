## **🟢 Part 1: Loan Eligibility & Interest Rate Calculation [4 marks]**  

📌 **Requirement:**  
SecureBank only approves loans if the **customer meets all eligibility criteria**. The system should:  
✔ **Check if the customer is employed.**  
✔ **Verify if the income is at least PKR 50,000.**  
✔ **Check credit score:**  
   - **750+ → 5% Interest Rate**  
   - **650 - 749 → 8% Interest Rate**  
   - **Below 650 → Loan Rejected**  
✔ **If the applicant is unemployed, the loan is rejected immediately.**  

### **Implementation Steps:**  
🔹 **Step 1:** Structure the `if-elif-else` statements for this logic.  
🔹 **Step 2:** Implement a Python program using **ipywidgets** for interactive user input (income, credit score, employment status) to decide whether the loan is approved or rejected.

---

```python
import ipywidgets as widgets
from IPython.display import display

def check_loan(employed, income, credit):
    if employed == 'No':
        return "Loan Rejected: Not employed"
    if income < 50000:
        return "Loan Rejected: Income < PKR 50,000"
    if credit >= 750:
        return "Approved! Rate: 5%"
    elif 650 <= credit < 750:
        return "Approved! Rate: 8%"
    else:
        return "Rejected: Credit score < 650"

employment = widgets.Dropdown(options=['Yes', 'No'], description='Employed:')
income = widgets.IntText(description='Income (PKR):')
credit = widgets.IntSlider(min=300, max=850, value=700, description='Credit Score:')
result = widgets.Output()

def on_check(_):
    with result:
        result.clear_output()
        output = check_loan(employment.value, income.value, credit.value)
        print(output)

check_btn = widgets.Button(description="Check Eligibility")
check_btn.on_click(on_check)


display(employment, income, credit, check_btn, result)
```

## **🟢 Part 2: Investment Risk Assessment [4 marks]**  

📌 **Requirement:**  
SecureBank offers **investment analysis services**. The system should evaluate a **portfolio of stocks** and classify the risk level based on stock returns:  
✔ If **any stock has a negative return**, the portfolio is **High Risk**.  
✔ If **all stocks have positive returns**, but at least one return is below 5%, classify as **Medium Risk**.  
✔ If **all stock returns are 5% or above**, classify as **Low Risk**.  

### **Implementation Steps:**  
🔹 **Step 1:** Use loops to iterate through the list of stock returns.  
🔹 **Step 2:** Implement `if-elif` conditions to classify the risk.  
🔹 **Step 3:** Write a Python program using **ipywidgets** to allow users to **input stock returns interactively** and receive a risk assessment.

---

```python
import ipywidgets as widgets
from IPython.display import display

def check_investment_risk(stock_percentages):
    for return_percent in stock_percentages:
        if return_percent < 0:
            return "High Risk - We found losing investments"

    for return_percent in stock_percentages:
        if return_percent < 5:
            return "Medium Risk - Some investments aren't growing enough"

    return "Low Risk - Your investments are growing well!"

stock_input = widgets.Text(
    description='Enter stock percentages:',
    placeholder='Example: 4, -1, 6',
    style={'description_width': '160px'}
)

check_button = widgets.Button(
    description='Check My Risk',
    button_style='primary'
)

result_display = widgets.Output()

def when_clicked(button):
    with result_display:
        result_display.clear_output()

        try:
            numbers = [float(n.strip()) for n in stock_input.value.split(',')]
            risk_message = check_investment_risk(numbers)
            print(f"🔍 Analysis Result: {risk_message}")

        except:
            print("⚠️ Please enter numbers like: 5, 3.5, -2")

check_button.on_click(when_clicked)

display(widgets.VBox([stock_input, check_button, result_display]))
```

## **🟢 Part 3: Loan Repayment Tracker [4 marks]**  

📌 **Requirement:**  
Customers who receive a loan should be able to track their **loan balance** as they make monthly payments. The system should:  
✔ Start with an **initial loan balance** (e.g., PKR 500,000).  
✔ Deduct a **fixed monthly payment** (e.g., PKR 25,000).  
✔ Continue tracking until **loan balance reaches zero**.  
✔ Display the remaining balance **after each payment**.  

### **Implementation Steps:**  
🔹 **Step 1:** Choose an appropriate loop (`for` or `while`).  
🔹 **Step 2:** Ensure the loop **stops once the loan is fully paid**.  
🔹 **Step 3:** Implement a Python program using **ipywidgets** to **simulate loan repayment interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display

def show_payment_plan(total_loan, monthly_amount):
    remaining = total_loan
    month_count = 1
    payment_history = []

    while remaining > 0:
        payment = monthly_amount if remaining >= monthly_amount else remaining
        remaining -= payment
        payment_history.append(f"Month {month_count}: Payment PKR {payment:,.2f} ➔ Remaining: PKR {remaining:,.2f}")
        month_count += 1

    return payment_history

loan_input = widgets.FloatText(
    value=500000,
    description="Total Loan Amount:",
    style={'description_width': '160px'}
)

payment_input = widgets.FloatText(
    value=25000,
    description="Monthly Payment:",
    style={'description_width': '160px'}
)

calculate_button = widgets.Button(
    description="Show My Payment Plan",
    button_style='success'
)

plan_display = widgets.Output()

def calculate_plan(_):
    with plan_display:
        plan_display.clear_output()

        try:
            if loan_input.value <= 0 or payment_input.value <= 0:
                print("🔔 Please enter amounts greater than zero")
                return

            if payment_input.value > loan_input.value:
                print("⚠️ Your monthly payment is larger than the loan amount!")
                return

            schedule = show_payment_plan(loan_input.value, payment_input.value)
            print("📅 Your Repayment Journey:")
            for step in schedule:
                print(step)

        except:
            print("❌ Oops! Please check your numbers")

calculate_button.on_click(calculate_plan)

display(widgets.VBox([loan_input, payment_input, calculate_button, plan_display]))
```


## **🟢 Part 4: Stock Price Monitoring and Trading Strategy [4 marks]**  

📌 **Requirement:**  
A stock trader wants to **track stock prices daily** and **sell when the price reaches PKR 200**. The system should:  
✔ Iterate through a **list of stock prices**.  
✔ **Skip missing stock data** (`None` values).  
✔ Stop tracking **once the price reaches PKR 200**.  

### **Implementation Steps:**  
🔹 **Step 1:** Handle missing stock data using `continue`.  
🔹 **Step 2:** Stop tracking once the stock hits the target price (`break`).  
🔹 **Step 3:** Write a Python program using **ipywidgets** to **process stock prices interactively** and trigger alerts when conditions are met.

---

```python
import ipywidgets as widgets
from IPython.display import display

def check_prices(daily_prices):
    found_target = False
    for day_number, price in enumerate(daily_prices, start=1):
        if price is None:
            yield f"Day {day_number}: 📭 Missing price data"
            continue

        yield f"Day {day_number}: ₹{price} - {'⚠️ Watching' if price < 200 else '🚨 Target reached!'}"

        if price >= 200:
            found_target = True
            break

    if not found_target:
        yield "✅ Monitoring ended: Price never reached ₹200"

price_entry = widgets.Text(
    placeholder='Example: 180, none, 190, 210',
    description='Daily Prices:',
    style={'description_width': '140px'}
)

analyze_button = widgets.Button(
    description='Check Prices Now',
    button_style='info'
)

report_display = widgets.Output()

def show_report(_):
    with report_display:
        report_display.clear_output()

        try:
            raw_input = price_entry.value.split(',')
            clean_prices = []

            for item in raw_input:
                clean_item = item.strip().lower()
                if clean_item == 'none':
                    clean_prices.append(None)
                else:
                    clean_prices.append(float(clean_item))

            print("📋 Your Stock Report:")
            for update in check_prices(clean_prices):
                print(update)

        except:
            print("❌ Oops! Please use numbers or 'none' like: 180, none, 200")

analyze_button.on_click(show_report)

display(widgets.VBox([price_entry, analyze_button, report_display]))
```


## **🟢 Part 5: Currency Exchange Rate Tracker [4 marks]**  

📌 **Requirement:**  
SecureBank provides **real-time currency exchange tracking**. The system should:  
✔ Start at **PKR 290/USD**.  
✔ Increase by **1 PKR per day** until it reaches **PKR 300/USD**.  
✔ Print exchange rates daily and stop when the **target rate is reached**.  

### **Implementation Steps:**  
🔹 **Step 1:** Choose a suitable loop (`for` or `while`).  
🔹 **Step 2:** Stop the loop when the exchange rate reaches the target.  
🔹 **Step 3:** Implement a Python program using **ipywidgets** to **track the currency exchange rate interactively**.

---

```python
import ipywidgets as widgets
from IPython.display import display

def track_exchange_rate():
    current_value = 290
    day = 1
    updates = []

    while current_value <= 300:
        updates.append(f"🌍 Day {day}: {current_value}pkr per USD")
        if current_value == 300:
            break
        current_value += 1
        day += 1

    return updates

track_button = widgets.Button(
    description='Track Now',
    button_style='primary',
    icon='money'
)

result_display = widgets.Output()

def show_progress(_):
    with result_display:
        result_display.clear_output()
        print("📅 Daily Exchange Rate Journey:")
        for update in track_exchange_rate():
            print(update)
        print("\n🎉 Success! Reached target rate of 300pkr/USD")

track_button.on_click(show_progress)

display(widgets.VBox([track_button, result_display]))
```
